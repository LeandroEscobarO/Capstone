```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
CoastLine_Lat = 28.56301
Coastline_lon = -80.56795
launch_sites_lat = 28.56321
launch_sites_lon = -80.57686
distance_coastline = calculate_distance(launch_sites_lat, launch_sites_lon, CoastLine_Lat, Coastline_lon)
distance_coastline
```

```python
distance_marker = folium.Marker(
    [CoastLine_Lat, Coastline_lon], # Changed to the coastline's coordinates
    icon=DivIcon(
        icon_size=(20,20),
        icon_anchor=(0,0),
        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance_coastline), # Using distance_coastline instead of distance
        )
    )
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
lines=folium.PolyLine(locations=[[launch_sites_lat, launch_sites_lon], [CoastLine_Lat, Coastline_lon]], weight=1) # Changed locations to a list of coordinate pairs
site_map.add_child(lines)
```
